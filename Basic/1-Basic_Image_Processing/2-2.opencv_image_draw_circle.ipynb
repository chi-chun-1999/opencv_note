# 基本影像繪圖 - 圓形、橢圓形
---

### 第一步、載入Python套件

```python
# 載入套件
import numpy as np
import cv2 as cv
```

### 第二步、創建一全黑的影像來作畫（或可以使用Lenna作畫）
<img src="img/attachment/Draw_empty.png" alt="Draw_empty" align="left">

```python
# 建立一張256x256的 RGB 圖片
# 函式輸出入 => 影像像素值矩陣 = np.zeros((高, 寬, 通道), 像素的資料型態)
img = np.zeros((256, 256, 3), np.uint8)
```

### 第三步、在影像上分別畫上圓形、橢圓形

```python
# 在影像上畫一圓形
# 函式輸出入 => 輸出影像 = cv.circle（輸入同輸出影像, 圓心座標, 半徑, 顏色, 線條寬度（-1為填滿））
img = cv.circle(img, (90, 210), 30, (0, 255, 255), -1)

# 在影像上畫一橢圓形
# 函式輸出入 => 輸出影像 = cv.ellipse（輸入同輸出影像, 中心點座標, 軸長, 旋轉角度, 起始角度, 結束角度, 顏色, 線條寬度（-1為填滿））
img = cv.ellipse(img, (180, 200), (25, 55), 45, 0, 360, (205, 0, 255), 2) 
```

### 第四步、開啟視窗並將影像顯示出來，按任意鍵將視窗關閉
<img src="img/attachment/Draw_circle.png" alt="Draw_circle" align="left">

```python
# 顯示繪圖影像
# 函式輸出入 => cv.imshow(視窗名稱, 影像矩陣)
cv.imshow('Draw circle/ellipse', img)

# 暫停程式並在按下任意鍵後結束暫停
# 函式輸出入 => 按鍵字元編碼 = cv.waitKey(停等時間單位毫秒)
cv.waitKey(0)

# 關閉所有OpenCV的視窗
# 函式輸出入 => cv.destroyAllWindows()
cv.destroyAllWindows()
```
